```python
import requests
from bs4 import BeautifulSoup
import os 
import re
import json
```

```python

caminhos = [os.path.join("ex1/paginas", nome) for nome in os.listdir("ex1/paginas")]
arquivos = [arq for arq in caminhos if os.path.isfile(arq)]
htmls = [arq for arq in arquivos if arq.lower().endswith(".html")]

for html in htmls:
    with open(html) as file:
        conteudo = ""
        for string in file.readlines():
            conteudo += string
        soup = BeautifulSoup(conteudo, "html.parser")
        print(soup.title.string)
        
        # verificar se tem uma infobox
        box = soup.select(".infobox")
        if len(box) > 0:
            # extraindo informações da infobox
            infomarcoes = {}
            box = soup.select(".infobox")[0]
            dados = {}
            if box.find("span"):
                #extraindo o titulo do box
                dados["titulo"] = box.find("span").text
            
            # para cada linha da infobox    
            for tag in box.find_all("tr"):
                chaves_extraidas = []
                valores_extraidos = []
                lista_valores = []
                
                if tag.find_all(attrs={"scope" : "row"}):
                    td_tag = tag.find_all("td")
                    try:
                        chaves = [filho for filho in td_tag[0].children]
                        valor = [filho for filho in td_tag[1].children]
                    except IndexError:
                        print("IndexError: td_tag does not have enough elements.")
                        continue
                    
                    for chave in chaves:
                        texto_chave = chave.text.strip()
                        if texto_chave:  
                            chaves_extraidas.append(texto_chave)
                            
                    for v in valor:
                        if re.search(r"<s*ul\s*", str(v)):
                            print("tem ul")
                            print(v)
                            lista_valores.append(v)
                            print("LISTA")
                            print(lista_valores)
                        else:
                            texto_valor = v.text.strip()
                            if texto_valor:
                                valores_extraidos.append(texto_valor)
                    
                    if len(chaves_extraidas) == 1:
                        chave_final = chaves_extraidas[0]  
                    else:
                        chave_final = tuple(chaves_extraidas)  

                    if lista_valores:
                        print(lista_valores)
                        valor_final = tuple([li.text.strip() for div in lista_valores for li in div.find_all("li")]
)
                    else:
                        valor_final = " ".join(valores_extraidos)

                    dados[chave_final] = valor_final
            nome_arquivo = os.path.join("jsons", f"{soup.title.string}.json")
            dados_json = {", ".join(k) if isinstance(k, tuple) else k: v for k, v in dados.items()}
            
            with open(nome_arquivo, "w", encoding="utf-8") as f:
                json.dump(dados_json, f, indent=4, ensure_ascii=False)

        else:
            print("NÃO tem uma infobox") 
        
```
